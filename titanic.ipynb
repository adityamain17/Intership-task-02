# EDA

# 1.What is the age distribution of passengers?

```python
# Plot age distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Age"],bins=30,kde=True,color="blue")

plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

# Gender Distribution 

```python
sns.barplot(df['Sex'])
```

# What is the distribution of fares paid?

```python
sns.boxplot(x="Sex", y="Fare", data=df)
plt.title("Fare Distribution by Sex")
plt.show()
```

# Number of Survivers

```python
df['Survived'].value_counts()
```

```python
survival_rate = df["Survived"].mean() * 100
print(f"Overall Survival Rate: {survival_rate:.2f}%")
```

# survival by gender

```python
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

# Survival By Pclass

```python
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

# Survival By Pclass And Sex

```python
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=df)
plt.title("Survival Rate by Class and Gender")
plt.show()
```

# Survival By Age Group 

```python
df["AgeGroup"] = pd.cut(df["Age"], bins=[0,12,18,40,60,80], 
                        labels=["Child","Teen","Adult","Middle-aged","Senior"])
df.groupby("AgeGroup")["Survived"].mean() * 100
```

```python
# Compute correlation matrix (only numeric columns)
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)

plt.title("Correlation Heatmap - Titanic Dataset")
plt.show()
```

# Conclusion

Overall Survival

Only about 38% of passengers survived.

Gender and Survival

Females had a much higher survival rate than males, supporting the women and children first policy.

Passenger Class and Survival

First-class passengers survived more often than second- and third-class passengers.

Third-class passengers had the lowest survival rate, showing the strong link between socio-economic status and survival chances.

Age Factor

Children (under 12) had higher survival rates compared to adults.

Older passengers had lower survival chances.

Fare and Survival

Passengers who paid higher fares (wealthier individuals, often in 1st class) had better survival chances.

Embarkation Point

Passengers who boarded from Cherbourg (C) had better survival rates compared to Southampton (S) and Queenstown (Q).

Family Effect

Passengers with small families (13 members) had slightly higher survival chances compared to those traveling alone or in very large families.

Correlation Insights (from heatmap)

Survived shows a positive correlation with Fare and Sex (female coded as 1).

Survived shows a negative correlation with Pclass (higher class  better survival).
